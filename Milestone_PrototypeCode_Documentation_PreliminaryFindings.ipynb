# Input and Modifying the Data

Here, I modified the data to seperate the systolic and diastolic blood pressure. In addition, I did some data cleaning by replacing null values in Sleep Disorder and BMI Category with "None" or uniformly making "Normal Weight" into "Normal."

The dataset comes from: https://data.mendeley.com/datasets/46j8wrc7p7/1. On Google Colab, you must upload it in the file repository but you may modify the file path for your needs.

```python
filepath = "/content/Sleep_health_and_lifestyle_dataset.csv"
```

```python
df = pd.read_csv(filepath)
df['Sleep Disorder'] = (df['Sleep Disorder'].replace(np.nan, 'None'))
df['BMI Category'] = (df['BMI Category'].replace('Normal Weight', 'Normal'))
bloodPressure = df['Blood Pressure'].str.split(pat='/', expand=True)
df['Systolic Blood Pressure'] = bloodPressure[0]
df['Diastolic Blood Pressure'] = bloodPressure[1]
df.head()
```

# Prototype Code

## Prototype Code

```python
# We will have Sleep Duration avs. Sleep Quality Blood Pressure, BMI, Heart Rate, Physical Activity Level - able to be sperated by Gender, Occupation, and Sleep Disorder as options

y_var = 'Quality of Sleep:Q'
x_var = 'Sleep Duration:Q'
y = 'Quality of Sleep'
x = 'Sleep Duration'

sq_width = 500
sq_height = 500
tip = ['Gender:N','Age:Q', 'Occupation:N','Sleep Duration:Q', 'Daily Steps:Q', 'Physical Activity Level:Q', 'BMI Category:N','Stress Level:Q']

sleepDisorder_select = alt.selection_point(fields=['Sleep Disorder'])
color_sleepDisorder= (alt.when(sleepDisorder_select).then(alt.value("lightgreen")).otherwise(alt.value("red")))

legend = alt.Chart(df).mark_rect().encode(
    alt.X('Sleep Disorder'),
    color = color_sleepDisorder,
    tooltip=alt.Tooltip(['count()'])
).add_params(sleepDisorder_select)

dropdown_vs_occ_select = alt.selection_point(fields=['Occupation'])
color_occ= (alt.when(dropdown_vs_occ_select).then(alt.value("lightgreen")).otherwise(alt.value("red")))

legend2 = alt.Chart(df).mark_rect().encode(
    alt.X('Occupation'),
    color = color_occ,
    tooltip=alt.Tooltip(['count()'])
).add_params(dropdown_vs_occ_select)


bmi_select = alt.selection_point(fields=['BMI Category'])
color_bmi= (alt.when(bmi_select).then(alt.value("lightgreen")).otherwise(alt.value("red")))

legend3 = alt.Chart(df).mark_rect().encode(
    alt.X('BMI Category:N'),
    color = color_bmi,
    tooltip=alt.Tooltip(['count()'])
).add_params(
    bmi_select
)

gender_select = alt.selection_point(fields=['Gender'])
color_gender= (alt.when(gender_select).then(alt.value("lightgreen")).otherwise(alt.value("red")))

legend4 = alt.Chart(df).mark_rect().encode(
    alt.X('Gender:N'),
    color = color_gender,
    tooltip=alt.Tooltip(['count()'])
).add_params(
    gender_select
)

points = alt.Chart(df).encode(
    alt.X(x_var),
    alt.Y(y_var),
    tooltip=alt.Tooltip(tip)
).properties(width=sq_width, height=sq_height).interactive().add_params(
    dropdown_vs_occ_select,bmi_select,sleepDisorder_select, gender_select
).transform_filter(
    dropdown_vs_occ_select,bmi_select,sleepDisorder_select, gender_select
)

all_points = alt.Chart(df).add_params(
    dropdown_vs_occ_select,bmi_select,sleepDisorder_select,gender_select
).transform_filter(
    dropdown_vs_occ_select,bmi_select,sleepDisorder_select,gender_select
)

chart_bp_a = all_points.encode(y='mean(Systolic Blood Pressure):Q', x='Quality of Sleep:Q').properties(width=300, height=300).mark_bar(color='red', width=35)
chart_bp_b = all_points.encode(y='mean(Diastolic Blood Pressure):Q', x='Quality of Sleep:Q').properties(width=300, height=300).mark_bar(color='blue', width=35)
chart_bp = chart_bp_a | chart_bp_b

chart2 = all_points.encode(y='mean(Physical Activity Level):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_bar(color='black', width=15)
chart3 = all_points.encode(y='mean(Stress Level):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_bar(color='black', width=15)
chart4 = all_points.encode(y='mean(Heart Rate):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_bar(color='black', width=15)
chart5 = all_points.encode(y='mean(Daily Steps):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_bar(color='black', width=15)
chart6 = all_points.encode(y='mean(Sleep Duration):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_bar(color='black', width=15)

layer1 =  (chart3 | chart4 | chart5 | chart6 | chart2)
layer1b = (chart_bp)
heatmap1 = alt.Chart(df).encode(
    alt.X('BMI Category'),
    alt.Y('Occupation'),
    color = alt.Color('count()').scale(scheme='purples').legend(None),
    tooltip=alt.Tooltip(['BMI Category','Occupation','count()'])
).properties(width=50, height=100).mark_rect()


heatmap2 = alt.Chart(df).encode(
    alt.X('Sleep Disorder'),
    alt.Y('Occupation'),
    color = alt.Color('count()').scale(scheme='purples').legend(None),
    tooltip=alt.Tooltip(['Sleep Disorder','Occupation','count()'])
).properties(width=50, height=100).mark_rect()


heatmap3 = alt.Chart(df).encode(
    alt.X('Sleep Disorder'),
    alt.Y('BMI Category'),
    color = alt.Color('count():Q').scale(scheme='purples').legend(None),
    tooltip=alt.Tooltip(['Sleep Disorder','BMI Category','count()'])
).properties(width=50, height=100).mark_rect()

heatmap4 = alt.Chart(df).encode(
    alt.X('Gender'),
    alt.Y('Sleep Disorder'),
    color = alt.Color('count():Q').scale(scheme='purples').legend(None),
    tooltip=alt.Tooltip(['Gender','Sleep Disorder','count()'])
).properties(width=50, height=100).mark_rect()

heatmap5 = alt.Chart(df).encode(
    alt.X('Gender'),
    alt.Y('Occupation'),
    color = alt.Color('count():Q').scale(scheme='purples').legend(None),
    tooltip=alt.Tooltip(['Gender','Occupation','count()'])
).properties(width=50, height=100).mark_rect()

layer2 = heatmap1 | heatmap2 | heatmap5| heatmap3 | heatmap4
layer3 = ((legend | legend2 | legend3 | legend4))
allLayers = alt.vconcat(layer1, layer1b, layer3,layer2, center = True).properties(title=alt.TitleParams(text='2024 Health and Sleep Relation Data from University of Oxford', subtitle='Dataset provided by Ben Cutting. Visualization by Risheek Chemudupati.')).configure(padding={'left': 50,'right': 50, 'top':20,'bottom':20})
allLayers.configure(autosize=alt.AutoSizeParams(resize=False))
allLayers.configure_title(fontSize=20, anchor="middle")
```

# Visualization

This is the code to display the whole visualization.

```python
layer2 = heatmap1 | heatmap2 | heatmap5| heatmap3 | heatmap4
layer3 = ((legend | legend2 | legend3 | legend4))
allLayers = alt.vconcat(layer1, layer1b, layer3,layer2, center = True).properties(title=alt.TitleParams(text='2024 Health and Sleep Relation Data from University of Oxford', subtitle='Dataset provided by Ben Cutting. Visualization by Risheek Chemudupati.')).configure(padding={'left': 50,'right': 50, 'top':20,'bottom':20})
allLayers.configure(autosize=alt.AutoSizeParams(resize=False))
allLayers.configure_title(fontSize=20, anchor="middle")
```

# Preliminary Report

## About The Dataset

This was based on a sleep disorder study funded by the University of Oxford to study sleep patterns with multiple data-types concerning age, heart rate, blood pressure, stress level, occupation, and more potentially interesting variables to compare, study, and explore. It was provided by Mendeley Data from Ben Cutting and published on March 20th, 2025.

## Data Cleaning

The total number of participants was 374. Some data-cleaning and processing was done to modify Normal Weight into Normal as well as null values in the Sleep Disorder category to None if it was null. No other values were imputed or replaced to not introduce skew to the results for the analysis, and outliers were not modified for the same reason.

## Descriptive Analysis

The average of the participants was 42.1 + 8.6 years old with a minimum age of 27 and a maximum age of 59. It seems that the average sleep duration was 7.1 hours.  So this would suggest that most participants were roughly middle-aged. The physical activity level was on average 59.17. There was a roughly even gender split, about 189 males and 185 females.  The occupation of the individuals varied with only one manager but also 73 nurses, 71 doctors, and 63 engineers. Most had no sleep disorder (count = 219), and less than 200 individuals had either sleep apnea (count = 78) or insomnia (count = 77).  

## Visualization Analysis

The exploratory visualization revealed that 61 Nurses had sleep apnea, given that 73 nurses were female. There were no males in the following occupations of scientist, nurse, and manager. There were no females in the occupations of sales representative, salesperson, or software engineer. Furthermore, there were 200 individuals who had normal BMI and no sleep disorder. There were 137 males with no sleep disorder compared to 82 females with no sleep disorder. None of the obese individuals had no sleep disorder and this was seen to a lesser degree with those who were overweight. Among those with no sleep disorder, regardless of occupation, BMI, and gender  they tended to have slightly decreasing mean heart rate or increasing mean sleep duration with increasing quality of sleep. Among those with sleep apnea, there was a large spike in mean physical activity, mean daily steps.  This was the same if they were female, were a nurse, or were overweight. Among males, in general, mean heart rate increased if they rated their quality of sleep a 9 compared to the overall trend of decreasing with increased quality of sleep. There was also a drop among this group for mean daily steps at a quality of sleep level of 9. This drop was still also noticed for those with sleep apnea.
