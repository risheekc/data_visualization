# Input and Modifying the Data

Here, I modified the data to seperate the systolic and diastolic blood pressure. In addition, I did some data cleaning by replacing null values in Sleep Disorder and BMI Category with "None" or uniformly making "Normal Weight" into "Normal."

The dataset comes from: https://data.mendeley.com/datasets/46j8wrc7p7/1. On Google Colab, you must upload it in the file repository but you may modify the file path for your needs.

```python
df = pd.read_csv(filepath)
df['Sleep Disorder'] = (df['Sleep Disorder'].replace(np.nan, 'None'))
df['BMI Category'] = (df['BMI Category'].replace('Normal Weight', 'Normal'))
bloodPressure = df['Blood Pressure'].str.split(pat='/', expand=True)
df['Systolic Blood Pressure'] = bloodPressure[0]
df['Diastolic Blood Pressure'] = bloodPressure[1]
df.head()
```

# Final Code

## Code

```python
# We will have Sleep Duration avs. Sleep Quality Blood Pressure, BMI, Heart Rate, Physical Activity Level - able to be sperated by Gender, Occupation, and Sleep Disorder as options

y_var = 'Quality of Sleep:Q'
x_var = 'Sleep Duration:Q'
y = 'Quality of Sleep'
x = 'Sleep Duration'

sq_width = 500
sq_height = 500
tip = ['Gender:N','Age:Q', 'Occupation:N','Sleep Duration:Q', 'Daily Steps:Q', 'Physical Activity Level:Q', 'BMI Category:N','Stress Level:Q']

sleepDisorder_select = alt.selection_point(fields=['Sleep Disorder'])
color_sleepDisorder= (alt.when(sleepDisorder_select).then(alt.value("lightgreen")).otherwise(alt.value("red")))

legend = alt.Chart(df).mark_rect().encode(
    alt.X('Sleep Disorder'),
    color = color_sleepDisorder,
    tooltip=alt.Tooltip(['count()'])
).add_params(sleepDisorder_select)

dropdown_vs_occ_select = alt.selection_point(fields=['Occupation'])
color_occ= (alt.when(dropdown_vs_occ_select).then(alt.value("lightgreen")).otherwise(alt.value("red")))

legend2 = alt.Chart(df).mark_rect().encode(
    alt.X('Occupation'),
    color = color_occ,
    tooltip=alt.Tooltip(['count()'])
).add_params(dropdown_vs_occ_select)


bmi_select = alt.selection_point(fields=['BMI Category'])
color_bmi= (alt.when(bmi_select).then(alt.value("lightgreen")).otherwise(alt.value("red")))

legend3 = alt.Chart(df).mark_rect().encode(
    alt.X('BMI Category:N'),
    color = color_bmi,
    tooltip=alt.Tooltip(['count()'])
).add_params(
    bmi_select
)

gender_select = alt.selection_point(fields=['Gender'])
color_gender= (alt.when(gender_select).then(alt.value("lightgreen")).otherwise(alt.value("red")))

legend4 = alt.Chart(df).mark_rect().encode(
    alt.X('Gender:N'),
    color = color_gender,
    tooltip=alt.Tooltip(['count()'])
).add_params(
    gender_select
)

points = alt.Chart(df).encode(
    alt.X(x_var),
    alt.Y(y_var),
    tooltip=alt.Tooltip(tip)
).properties(width=sq_width, height=sq_height).interactive().add_params(
    dropdown_vs_occ_select,bmi_select,sleepDisorder_select, gender_select
).transform_filter(
    dropdown_vs_occ_select,bmi_select,sleepDisorder_select, gender_select
)

all_points = alt.Chart(df).add_params(
    dropdown_vs_occ_select,bmi_select,sleepDisorder_select,gender_select
).transform_filter(
    dropdown_vs_occ_select,bmi_select,sleepDisorder_select,gender_select
)

chart2 = all_points.encode(y='mean(Physical Activity Level):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart3 = all_points.encode(y='mean(Stress Level):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart4 = all_points.encode(y='mean(Heart Rate):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart5 = all_points.encode(y='mean(Daily Steps):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart6 = all_points.encode(y='mean(Sleep Duration):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart7 = all_points.encode(y='mean(Diastolic Blood Pressure):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart8 = all_points.encode(y='mean(Systolic Blood Pressure):Q', x='Quality of Sleep:Q').properties(width=100, height=100).mark_line(color='black', width=15)

heatmap_1 = alt.Chart(df).encode(
    alt.X('Quality of Sleep:N'),
    alt.Y('Physical Activity Level:Q'),
    color = alt.Color('count():Q').legend(None),
    tooltip=alt.Tooltip(['Quality of Sleep','Physical Activity Level','count()'])
).properties(width=50, height=100).mark_rect()

heatmap_2 = alt.Chart(df).encode(
    alt.X('Quality of Sleep:N'),
    alt.Y('Stress Level:Q'),
    color = alt.Color('count():Q').legend(None),
    tooltip=alt.Tooltip(['Quality of Sleep','Stress Level','count()'])
).properties(width=50, height=100).mark_rect()

heatmap_3 = alt.Chart(df).encode(
    alt.X('Quality of Sleep:N'),
    alt.Y('Heart Rate:Q'),
    color = alt.Color('count():Q').legend(None),
    tooltip=alt.Tooltip(['Quality of Sleep','Heart Rate','count()'])
).properties(width=50, height=100).mark_rect()

heatmap_4 = alt.Chart(df).encode(
    alt.X('Quality of Sleep:N'),
    alt.Y('Sleep Duration:Q'),
    color = alt.Color('count():Q').legend(None),
    tooltip=alt.Tooltip(['Quality of Sleep','Sleep Duration','count()'])
).properties(width=50, height=100).mark_rect()

layer1 =  (heatmap_1 | heatmap_2 | heatmap_3 | heatmap_4)

chart9 = all_points.encode(y='mean(Physical Activity Level):Q', x='Sleep Duration:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart10 = all_points.encode(y='mean(Stress Level):Q', x='Sleep Duration:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart11 = all_points.encode(y='mean(Heart Rate):Q', x='Sleep Duration:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart12 = all_points.encode(y='mean(Daily Steps):Q', x='Sleep Duration:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart13 = all_points.encode(y='mean(Quality of Sleep):Q', x='Sleep Duration:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart14 = all_points.encode(y='mean(Diastolic Blood Pressure):Q', x='Sleep Duration:Q').properties(width=100, height=100).mark_line(color='black', width=15)
chart15 = all_points.encode(y='mean(Systolic Blood Pressure):Q', x='Sleep Duration:Q').properties(width=100, height=100).mark_line(color='black', width=15)

layer1 =  (chart2 | chart3 | chart4 | chart5 | chart6 | chart7 | chart8)
layer2 =  (chart9 | chart10 | chart11 | chart12 | chart13 | chart14 | chart15)

heatmap1 = alt.Chart(df).encode(
    alt.X('BMI Category'),
    alt.Y('Occupation'),
    color = alt.Color('count()').scale(scheme='purples').legend(None),
    tooltip=alt.Tooltip(['BMI Category','Occupation','count()'])
).properties(width=50, height=100).mark_rect()


heatmap2 = alt.Chart(df).encode(
    alt.X('Sleep Disorder'),
    alt.Y('Occupation'),
    color = alt.Color('count()').scale(scheme='purples').legend(None),
    tooltip=alt.Tooltip(['Sleep Disorder','Occupation','count()'])
).properties(width=50, height=100).mark_rect()


heatmap3 = alt.Chart(df).encode(
    alt.X('Sleep Disorder'),
    alt.Y('BMI Category'),
    color = alt.Color('count():Q').scale(scheme='purples').legend(None),
    tooltip=alt.Tooltip(['Sleep Disorder','BMI Category','count()'])
).properties(width=50, height=100).mark_rect()

heatmap4 = alt.Chart(df).encode(
    alt.X('Gender'),
    alt.Y('Sleep Disorder'),
    color = alt.Color('count():Q').scale(scheme='purples').legend(None),
    tooltip=alt.Tooltip(['Gender','Sleep Disorder','count()'])
).properties(width=50, height=100).mark_rect()

heatmap5 = alt.Chart(df).encode(
    alt.X('Gender'),
    alt.Y('Occupation'),
    color = alt.Color('count():Q').scale(scheme='purples').legend(None),
    tooltip=alt.Tooltip(['Gender','Occupation','count()'])
).properties(width=50, height=100).mark_rect()
```

# Visualization

This is the code to display the whole visualization.

```python
layer3 = ((legend | legend2 | legend3 | legend4))
layer4 = heatmap1 | heatmap2 | heatmap5| heatmap3 | heatmap4
allLayers = alt.vconcat(layer1, layer2, layer3,layer4, center = True).properties(title=alt.TitleParams(text='2024 Health and Sleep Relation Data from Mendeley Data', subtitle='Dataset provided by Ben Cutting. Visualization by Risheek Chemudupati.')).configure(padding={'left': 50,'right': 50, 'top':20,'bottom':20})
allLayers.configure(autosize=alt.AutoSizeParams(resize=False))
allLayers.configure_title(fontSize=20, anchor="middle")
```
